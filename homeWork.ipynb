```python
import random
import math
import tkinter as tk
tableOfColors = [
    "#FF5733",  # Red-Orange
    "#33FF57",  # Lime Green
    "#3357FF",  # Blue
    "#FFFF33",  # Yellow
    "#FF33FF",  # Magenta
    "#33FFFF",  # Cyan
    "#FF9933",  # Orange
    "#9933FF",  # Purple
    "#33FF99",  # Aqua Green
    "#FF6633",  # Coral
    "#6699FF",  # Light Blue
    "#99FF33",  # Yellow-Green
    "#FF3366",  # Rose Red
    "#33FF66",  # Spring Green
    "#6633FF",  # Violet
    "#FF3333",  # Bright Red
    "#33FF33",  # Neon Green
    "#663399",  # Rebecca Purple
    "#FF6600",  # Dark Orange
    "#00FF66"   # Electric Green
]
n= 8
defectueuesex= 6
defectueuesey = 5



def choisir_couleur(matrix, x, y, size, position, contain_defectueuse):
    print("x = ", x)
    print("y = ", y)
    print("size = ", size)
    """
    Fonction pour colorer la matrice en fonction de la position donnée.
    :param matrix: Matrice à colorer
    :param x: Coordonnée de départ pour les lignes
    :param y: Coordonnée de départ pour les colonnes
    :param size: Taille de la matrice actuelle
    :param position: Position du sous-bloc à colorer (top-left, top-right, bottom-left, bottom-right)
    """
    random_color = random.choice(tableOfColors)
    
    if position == 'top-left' or (position =='bottom-right' and (x>0 and x<n/2 -1 and y < n/2)):
            # Colorier top-left comme :
            # [c, c]
            # [c, .]
            matrix[x][y] = random_color
            matrix[x][y+1] = random_color
            matrix[x+1][y] = random_color
            matrix[x+1][y+1] = '.'
            
        
    elif position == 'bottom-left' or (position == 'top-right' and x>3 and y <n/2):
            # Colorier top-right comme :
            # [c, c]
            # [., c]
            matrix[x][y+1] = '.'
            matrix[x][y] = random_color
            matrix[x+1][y] = random_color
            matrix[x+1][y+1] = random_color
        
    
        
    elif position == 'top-right':
                    # Colorier bottom-left comme :
                    # [c, .]
                    # [c, c]
                    matrix[x][y] = random_color
                    matrix[x][y+1] = random_color
                    matrix[x+1][y] = '.'
                    matrix[x+1][y+1] = random_color
    
        
    elif position == 'bottom-right':
            # Colorier bottom-right comme :
            # [., c]
            # [c, c]
            matrix[x][y] = '.'
            matrix[x][y+1] = random_color
            matrix[x+1][y] = random_color
            matrix[x+1][y+1] = random_color
        
    
        


def carrelage_reccursif(matrix, x, y, size, position, contain_defectueuse):
    
    if n==4:
        print("simple")
    
    elif  (math.log2(n) % 1) != 0:
        print('not accepted')
        
    else :
        if size == 2:
            # Base case : call choisir_couleur when matrix is of size 2x2
            choisir_couleur(matrix, x, y, size, position, contain_defectueuse)
            return
        
        half = size // 2
        
        # Diviser la matrice en 4 quadrants
        carrelage_reccursif(matrix, x, y, half, "top-left",False)          # Quadrant haut-gauche
        carrelage_reccursif(matrix, x, y + half, half, 'top-right', False)  # Quadrant haut-droit
        carrelage_reccursif(matrix, x + half, y, half, 'bottom-left', False) # Quadrant bas-gauche
        carrelage_reccursif(matrix, x + half, y + half, half, 'bottom-right', False)
        

matrix = [['.' for _ in range(n)] for _ in range(n)]  # Initialiser une matrice vide de taille 8x8

# Appel de la fonction récursive
carrelage_reccursif(matrix, 0, 0, n, "all", False)

# Affichage de la matrice
for row in matrix:
    print(row)


import tkinter as tk

def create_color_matrix_gui(matrix, n=2, defectueuesex=None, defectueuesey=None):
    """
    Creates a GUI that displays a matrix of colors.

    :param matrix: A 2D list (matrix) of color strings (hexadecimal).
    :param n: The size of the matrix. Defaults to 2.
    :param defectueuesex: Row index of a specific defective cell, if any.
    :param defectueuesey: Column index of a specific defective cell, if any.
    """
    if n == 2:
        # Define a default 2x2 matrix
        matrix = [
            ["#00FF66", "#00FF66"],
            ["#00FF66", "."]
        ]

    # Create the main window
    root = tk.Tk()
    root.title("Color Matrix")

    # Loop through the matrix and create buttons for each cell
    for i, row in enumerate(matrix):
        for j, color in enumerate(row):
            # Handle defective cell coloring
            
            if(i==5 and j==6):
                color = "black"
            elif color == ".":
                color = "pink"

            # Create a button for each cell with the background set to the color
            button = tk.Button(
                root,
                bg=color ,
               
                width=8,
                height=4
            )
            button.grid(row=i, column=j, padx=2, pady=2)  # Place the button on the grid

    # Run the Tkinter event loop
    root.mainloop()

# Example usage

create_color_matrix_gui(matrix, n=3, defectueuesex=1, defectueuesey=1)

```
